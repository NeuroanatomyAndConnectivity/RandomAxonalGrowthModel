```python
import pandas as pd
import numpy as np
```

setting path to output text file from rag script (neurons and axons), here I saved several files appending the name of each run 

```python
path='/scr/ilz1/rag/'
run='1'
```

read file into pandas dataframe and split into arrays

```python
full_df=pd.read_csv(path+'ragv3naxons_%s.txt'%run, header=None)
```

```python
neurons=array(full_df.iloc[:,[0,1]])
```

```python
axons=array(full_df.iloc[:,[2,3]])
```

plotting initial neurons with filled circles, endpoints with open circles and lines in different colors, saves the plot in the same directory as the original textfile

```python
fig=figure(figsize=(8,8))
fig.suptitle('ragv3 #%s'%run, fontsize=20)
ax=fig.add_axes()
ax=plot((neurons[:,0], axons[:,0]), (neurons[:,1],axons[:,1]))#, marker='o')
ax=scatter(neurons[:,[0]], neurons[:,[1]], color='black', marker='o', s=40)
ax=scatter(axons[:,[0]], axons[:,[1]], color='black', marker='o', facecolor='none', s=40)
fig.savefig(path+'ragv3naxons_%s.png'%run)
```
